```python
class Gene:
    def __init__(self,subject,sec,teacher,room,time):
        self.subject = subject
        self.section = sec
        self.teacher = teacher
        self.room = room
        self.time = time
    
    def display(self):
        print(f'Class of {self.section} at {self.time} in {self.room} of subject {self.subject} teacher {self.teacher}')
```

```python
class Chromosome:
    def __init__(self,genes_list):
        self.genes_list = genes_list
        self.fitness_value = None  
        self.flag = False  # for soft constraints
        
    def soft_constraints(self):
        total_teacher_classes = {}
        total_student_classes = {}
        
        for gene in self.genes_list:
            
            if(gene.teacher not in total_teacher_classes.keys()): # counts total classes of all teachers
                total_teacher_classes[gene.teacher] = 1
            else:
                total_teacher_classes[gene.teacher] += 1
            
            if(gene.section not in total_student_classes.keys()): # counts total classes of all students
                total_student_classes[gene.section] = 1
            else:
                total_student_classes[gene.section] += 1
                    
        if (max(list(total_teacher_classes.values())) > 3): # if anyone teacher has more than 3 total classes
            self.fitness_value += 3 # give penalty
            
        if (max(list(total_student_classes.values())) > 4): # if anyone student has more than 4 total classes 
            self.fitness_value += 4 # give penalty

        if (max(total_teacher_classes.values()) <= 3 and max(total_student_classes.values()) <= 4): 
            self.flag = True  # if soft constraints are satisfied
        else:
            self.flag = False
    
    def calculate_fitness(self):
        time_vs_sec = {}
        time_vs_room = {}
        time_vs_teacher = {}
        clashes= 0
        
        for gene in self.genes_list:   
            
            if (gene.time not in time_vs_sec.keys()): # stores list of sections against corresponding time slots
                time_vs_sec[gene.time] = [gene.section]
            else:
                time_vs_sec[gene.time] = time_vs_sec[gene.time] + [gene.section]
            
            if (gene.time not in time_vs_room.keys()): # stores list of rooms against corresponding time slots
                time_vs_room[gene.time] = [gene.room]
            else:
                time_vs_room[gene.time] = time_vs_room[gene.time] + [gene.room]
                
            if (gene.time not in time_vs_teacher.keys()):  # stores list of teachers against corresponding time slots
                time_vs_teacher[gene.time] = [gene.teacher]
            else:
                time_vs_teacher[gene.time] = time_vs_teacher[gene.time] + [gene.teacher]
        
        for time in time_slots:
            
            room_list = []
            sec_list = []
            teacher_list = []
            
            #checking if more than 1 particular student/teacher/room is there in a particular time_slot
            
            if(time in time_vs_room.keys()):
                room_list = time_vs_room[time]
                
            if(time in time_vs_sec.keys()):
                sec_list = time_vs_sec[time]
            
            if(time in time_vs_teacher.keys()):
                teacher_list = time_vs_teacher[time]
                
            clashes = clashes + (len(room_list)-len(set(room_list))) + (len(sec_list)-len(set(sec_list))) + (len(teacher_list)-len(set(teacher_list)))
        
        self.fitness_value = clashes # lesser the fitness value, better the chromosome
        self.soft_constraints() 
    
    def display(self):
        for gene in self.genes_list:
            print(f'subject:{gene.subject} | section:{gene.section} | room:{gene.room} | time:{gene.time} | teacher:{gene.teacher}')
            
    def print_time_table(self):
        df = pd.DataFrame(columns=time_slots)
        df.insert(loc=0,column='Room',value=rooms)
        df.fillna('-',inplace=True) 

        for gene in self.genes_list:
            ind = np.where(df['Room'] == gene.room)[0][0]
            df.loc[ind,gene.time] = gene.subject + ',' + gene.section + ',' + gene.teacher  

        return df
```

```python
class Population:
    def __init__(self,chromosome_list):
        self.chromosomes_list = chromosome_list
        
    def cross_over(self,parent_1,parent_2): 
            
            cross_over_point = random.randint(0,len(parent_2.genes_list)-1)

            #till crossover-point, we get 1st part of child_1 & child_2
            child_1_gene_list = parent_1.genes_list[0:cross_over_point]  
            child_2_gene_list = parent_2.genes_list[0:cross_over_point]

 #for 2nd part of child_1 & child_2, swap room & time only, swapping other genes will result in same subjects multiple times
            for gene_1,gene_2 in zip(parent_1.genes_list[cross_over_point:],parent_2.genes_list[cross_over_point:]):

                child_1_gene = Gene(gene_1.subject,gene_1.section,gene_1.teacher,gene_2.room,gene_2.time)
                child_2_gene = Gene(gene_2.subject,gene_2.section,gene_2.teacher,gene_1.room,gene_1.time)

                child_1_gene_list.append(child_1_gene)
                child_2_gene_list.append(child_2_gene)

            child_1 = Chromosome(child_1_gene_list)
            child_2 = Chromosome(child_2_gene_list)
                
            child_1.calculate_fitness()
            child_2.calculate_fitness()
                
            return child_1,child_2
        
    def mutation(self,child,MUTATION_RATE):
        for i in range(4):

            if (random.random() < MUTATION_RATE):
            
                gene_no = random.randint(0,len(child.genes_list)-1) # randomly select a gene to change
                new_room = random.choice(rooms)   # randomly select room
                new_time = random.choice(time_slots)  # randomly select time
                rand_course = random.choice(subject) # randomly select course
                
                chromo_courses= []
                for i in child.genes_list: # get all courses in current chromosome
                    course = i.subject + ':' + i.section + ':' + i.teacher 
                    chromo_courses.append(course)
                    
                if(rand_course in chromo_courses):
                    while(1): # keep selecing a random course that is not already present in cureent chromosome
                        rand_course = random.choice(subject)
                        if(rand_course not in chromo_courses):
                            break
                    
                new_sub , new_sec, new_teacher = rand_course.split(':')

                child.genes_list[gene_no].subject = new_sub
                child.genes_list[gene_no].section = new_sec
                child.genes_list[gene_no].teacher = new_teacher
                child.genes_list[gene_no].room = new_room
                child.genes_list[gene_no].time = new_time
                    
                child.calculate_fitness()
                
        return child
    
    def torunament_selection(self,k=3):
        parent_1 = 0
        parent_2 = 0

        while(parent_1 == parent_2):  #keep selecting unitl two different parents are found
            counter = count() # for unique value else error is raised while comparing 
            
            sample_1 = random.sample(self.chromosomes_list, k) # select k different random chromosomes
            sample_2 = random.sample(self.chromosomes_list, k)
            
            fitness_list_1 = []
            fitness_list_2 = []
            
            for i in range(k):
# adding unique value in fitness list so that if 2 chromsomes have same fitness then they are sorted wrt to counter val because chromosomes can't be sorted
                unique_val = next(counter) 
                fitness_list_1.append((sample_1[i].fitness_value,unique_val,sample_1[i])) #(chromo.fitness_val,counter,chromo)
                fitness_list_2.append((sample_2[i].fitness_value,unique_val,sample_2[i])) 
            
            fitness_list_1.sort()
            fitness_list_2.sort()
            
            parent_1 = fitness_list_1[0][2] # select first 2 chromosome from each list
            parent_2 = fitness_list_2[0][2]
            
        return parent_1,parent_2
```

```python
def create_population(pop_size):
    
    chromosomes_list = []
    for x in range(0,pop_size):
        gene_list = []
        closed_list = []

        while(len(gene_list) != 150):
            
            course_no = random.randint(0,len(subject)-1)
            course = subject[course_no] #pick a random course(sub,sec,teacher) from subjects list
            
            if(course not in closed_list): # if course is not already bein added then add it
                closed_list.append(course)
                sub,sec,teacher = course.split(":")

                room = random.choice(rooms) # selecting random room
                time = random.choice(time_slots)  # selecting random time
                gene = Gene(sub,sec,teacher,room,time)
                gene_list.append(gene)

        single_chromosome = Chromosome(gene_list)
        single_chromosome.calculate_fitness()
        chromosomes_list.append(single_chromosome) 
        
    ppl = Population(chromosomes_list)
    return ppl
```

```python
def best_chromosome(population): # finding best chromo(with least fitness value) in current population
    fitness_values = [population.chromosomes_list[i].fitness_value for i in range(len(population.chromosomes_list))]
    fitness_values = np.array(fitness_values)
    index = fitness_values.argmin() # get index of chromosome with least fitness value
    best_chromo = population.chromosomes_list[index]
    
    print('best chromosome fitness value (No of clashes): ', best_chromo.fitness_value)
    if(best_chromo.fitness_value == 0): # if hard & soft constraints are satisfied
        print("Hard constraints satisfied")
        if(best_chromo.flag == True): # it will always be True - if not then fitness value can never be zero because of penalty
            print('Soft Constraints satisfied')
            print("SOLUTION FOUND")
            solution = best_chromo.print_time_table()
            return solution
```

```python
def Elitism(old_pop): # Select best 2 from old pop and transfer them into new pop
    fitness_old = [old_pop.chromosomes_list[i].fitness_value for i in range(len(old_pop.chromosomes_list))]
    
    fitness_old = np.array(fitness_old)
    
    fitness_old = fitness_old.argsort() # sorting all chromosomes' fitness values index wise

    chromo_1 = old_pop.chromosomes_list[fitness_old[0]]
    chromo_2 = old_pop.chromosomes_list[fitness_old[1]]
    
    return chromo_1,chromo_2
```

```python
def GA(MUTATION_RATE,NUM_OF_GENERATIONS,POPULATION_SIZE):

    population = create_population(POPULATION_SIZE)
    solution = None
    
    for generation in range(NUM_OF_GENERATIONS):

        print(f' Generation No: {generation+1}')

        solution = best_chromosome(population)
        if(type(solution)==pd.DataFrame): # if sol not found then None is returned else a df is returned
            return solution
        
        child_population = []

        for _ in range((POPULATION_SIZE//2)-2): #pop_size=50 => 50/2 = 25 => 25-2 => 23, other two come from elistism
            
            parent_1, parent_2 = population.torunament_selection()
            child_1, child_2 = population.cross_over(parent_1, parent_2)

            child_1 = population.mutation(child_1,MUTATION_RATE)
            child_2 = population.mutation(child_2,MUTATION_RATE)

            child_population.append(child_1)
            child_population.append(child_2)
        
        parent_1,parent_2 = Elitism(population)
        child_population.append(parent_1) 
        child_population.append(parent_2) 
        population = Population(child_population)
```

```python
MUTATION_RATE = 0.01
NUM_OF_GENERATIONS = 15000
POPULATION_SIZE = 50

solution = GA(MUTATION_RATE,NUM_OF_GENERATIONS,POPULATION_SIZE)
solution
```
